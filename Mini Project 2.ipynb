# Labour Training Evaluation Data Analysis
This notebook performs a comprehensive data analysis, visualization based on labour training evaluation data. It includes 17+ visual insights.
Note : This notebook only contain the visualization for data insights and doesn't contains any machine learning model or something like that

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv("LabourTrainingEvaluationData.csv")
df.head()
```

```python
# Checking missing values
print(df.isnull().sum())

# Basic info
df.info()

# Check for duplicates
df.duplicated().sum()

# Ensure proper data types
df['Eduacation'] = df['Eduacation'].astype('category')
df['Race'] = df['Race'].astype('category')
df['Hisp'] = df['Hisp'].astype('category')
df['MaritalStatus'] = df['MaritalStatus'].astype('category')
```

## Data Visualization (17 Graphs)

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(y='Eduacation', data=df, order=df['Eduacation'].value_counts().index)
plt.title('Education Level Count')
plt.show()
```

```python
sns.countplot(x='Race', data=df)
plt.title('Race Distribution')
plt.show()
```

```python
sns.countplot(x='Hisp', data=df)
plt.title('Hispanic Representation')
plt.show()
```

```python
sns.countplot(x='MaritalStatus', data=df)
plt.title('Marital Status Distribution')
plt.show()
```

```python
sns.countplot(x='Nodeg', data=df)
plt.title('No Degree Distribution')
plt.xticks([0,1], ['Degree', 'No Degree'])
plt.show()
```

```python
sns.histplot(df['Earnings_1974'], bins=50, kde=True)
plt.title('Earnings 1974 Distribution')
plt.show()
```

```python
sns.histplot(df['Earnings_1975'], bins=50, kde=True)
plt.title('Earnings 1975 Distribution')
plt.show()
```

```python
sns.histplot(df['Earnings_1978'], bins=50, kde=True)
plt.title('Earnings 1978 Distribution')
plt.show()
```

```python
df['Earnings_Change'] = df['Earnings_1978'] - df['Earnings_1974']
sns.histplot(df['Earnings_Change'], bins=50, kde=True)
plt.title('Change in Earnings (1978 - 1974)')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Earnings_1978', data=df)
plt.title('Age vs Earnings 1978')
plt.show()
```

```python
sns.boxplot(x='Eduacation', y='Earnings_1978', data=df)
plt.title('Education Level vs Earnings 1978')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='MaritalStatus', y='Earnings_1978', data=df)
plt.title('Marital Status vs Earnings 1978')
plt.show()
```

```python
sns.scatterplot(x='Earnings_1975', y='Earnings_1978', data=df)
plt.title('Earnings 1975 vs 1978')
plt.show()
```
